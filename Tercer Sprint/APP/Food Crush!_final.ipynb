```python
# Creamos la app
%%writefile Food_Crush.py
# Importamos librer铆as:
import streamlit as st
import lottie
import time
import streamlit_option_menu
from streamlit_option_menu import option_menu
import emoji
from PIL import Image

# Posici贸n correcta:
st.set_page_config(
    page_title="Food Crush!",
    layout="wide",
    initial_sidebar_state="expanded",
)

# Inicializar selected_option fuera del bloque with
selected_option = None
# Encabezado
st.header('Bienvenido a la mejor App para elegir comidas', divider='rainbow')
st.header('Encuentra tu _Match_ :grey[HOY] :sunglasses:')
# T铆tulo e 铆cono
st.title("Food :red[Crush]" + emoji.emojize(":hamburger:"))
# bot贸n interactivo
on = st.toggle('Modo hambre')
# Barra deslizadora
x = st.slider('Selecciona tus ganas de comer, algo rico')
if on:
    st.write('隆Modo hambre activado!')


# Crea la barra lateral
with st.sidebar:
    # Menu options
    selected_option = option_menu(
        menu_title="Men煤 principal:",
        options=["Inicio","Recomendaciones"], # "Estados", "Recomendaciones", "Restaurantes"],
        default_index=0,  # Set default option to "Inicio" (index 0)
        icons=["cast", "map", "pen", "book"],
        menu_icon="bars",
    )



def main():


    # Contenido de la p谩gina
    if selected_option == "Inicio":
        st.write("隆Hola! 驴Todav铆a no comiste? 驴Qu茅 esperas para aprovechar los descuentos? :sunglasses:")
        st.write("Aqu铆 puedes encontrar informaci贸n sobre los mejores restaurantes, bares y consejos para una experiencia inolvidable.")
        st.write("**Explora nuestras opciones:**")
        st.write("- **Recomendaciones:** Encuentra tu lugar ideal para deleitarte con comidas y tragos.")
        st.write("- **Restaurantes:** Descubre nuevos lugares para ir a comer :heart_eyes:.")
        st.write("- **Consejos:** Mejora tus lugares preferidos :wink:.")
        st.write("- **Descuentos:** Hasta un 60% OFF en restaurantes y beneficios exclusivos, para usuarios de Food Crush!")

    elif selected_option == "Recomendaciones":
        recomendaciones={'Pensilvania':["Domino's Pizza","Pizza Hut","Little Caesars Pizza","Papa John's Pizza","Burger King"],
                  'Ilinois':["Pizza Hut","Domino's Pizza","Casey's","Rosati's Pizza","Taco Bell"],
                  'Florida':["Papa John's Pizza","Pier 36 Fish Camp, Seafood,Oyster, Tapas,Bar & Grill","Domino's Pizza","The Breadboard","Harold's"],
                  'California':["Pizza Hut","Domino's Pizza","Doc's Drive In ","Burger King","Dickey's Barbecue Pit"]}



        selected_state = st.selectbox("Seleccione un estado", ["Pensilvania", "Ilinois","California","Florida"],key='Estado',index=None)
        if selected_state is not None:
            st.write(f"Restarurantes recomendados:\n")
            for restaurante in recomendaciones[selected_state]:
              st.write(restaurante)



if __name__ == "__main__":
    main()
```
